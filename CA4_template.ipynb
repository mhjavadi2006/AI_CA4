### Create a Node class

```python
class Node:
    def __init__(self, feature=None, threshold=None, left=None, right=None, value=None):
        # TODO: Store the feature name used for splitting
        # TODO: Store the threshold value
        # TODO: Store the left child
        # TODO: Store the right child
        # TODO: Store the final prediction value for leaf nodes

        self.feature = feature
        self.threshold = threshold
        self.left = left
        self.right = right
        self.value = value

    def is_leaf(self):
        # TODO: Return True if this node is a leaf
        if self.value is not None:
            return True

        return False
```

### Decision Tree Classifier from Scratch

```python
class DecisionTreeClassifierScratch:
    def __init__(self, max_depth=5, min_samples_split=20, min_samples_leaf=10):
        # TODO: Store max_depth
        # TODO: Store min_samples_split
        # TODO: Store min_samples_leaf
        # TODO: Create a variable named root and set it to None
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.min_samples_leaf = min_samples_leaf
        self.root = None

    def entropy(self, y):
        # TODO: Calculate entropy using class probabilities
        # Hint:
        # values, counts = np.unique(y, return_counts=True)
        # probabilities = counts / len(y)

        if len(y) == 0:
            return 0
        
        values, counts = np.unique(y,return_counts=True)

        probabilities = counts / len(y)

        return -np.sum(probabilities * np.log2(probabilities))

    def information_gain(self, y, y_left, y_right):
        # TODO: Calculate parent entropy
        # TODO: Calculate weighted child entropy
        # TODO: Return information gain
        
        H_s = self.entropy(y)

        H_l = self.entropy(y_left)

        H_r = self.entropy(y_right)

        weight_l = len(y_left) / len(y)
        weight_r = len(y_right) / len(y)

        IG = H_s - weight_r * H_r - weight_l * H_l

        return IG

    def majority_class(self, y):
        # TODO: Return the most common class in y
        if len(y) == 0:
            return 0
        

        values, counts = np.unique(y,return_counts=True)

        best_idx = np.argmax(counts)

        return values[best_idx]
    
    
    def find_best_split(self, X, y):
        # TODO: Try each feature
        # TODO: Try possible thresholds for each feature
        # TODO: Split y into left and right groups
        # TODO: Calculate information gain
        # TODO: Return best feature and best threshold
        
        best_ig = -1

        best_feature = None

        best_threshold = None

        samples, features = X.shape

        for f in range(features):

            X_c = X[:,f]

            thresholds = np.unique(X_c)

            for t in thresholds:
                left = X_c <= t
                right = X_c > t

                y_left = y[left] 
                y_right = y[right]

                if len(y_left) < self.min_samples_leaf or len(y_right) < self.min_samples_leaf:
                    continue

                ig = self.information_gain(y,y_left,y_right)

                if ig > best_ig:
                    best_ig = ig
                    best_feature = f
                    best_threshold = t

        return best_feature,best_threshold

        
    def build_tree(self, X, y, depth=0):
        # TODO: Check stopping conditions
        # TODO: Find the best split
        # TODO: Build left and right child nodes recursively
        # TODO: Return a Node

        samples, features = X.shape

        if depth >= self.max_depth or samples < self.min_samples_split or len(np.unique(y)) == 1:
            leaf_value = self.majority_class(y)

            return Node(value= leaf_value)

        best_feature, best_threshold = self.find_best_split(X,y)
        
        if best_feature is None:
            leaf_value = self.majority_class(y)
            return Node(value=leaf_value)
        left = X[:,best_feature] <= best_threshold
        right = X[:,best_feature] > best_threshold
        
        left_child = self.build_tree(X[left],y[left],depth + 1)
        right_child = self.build_tree(X[right],y[right],depth + 1)

        return Node(left=left_child,right=right_child,feature=best_feature,threshold=best_threshold)

    def fit(self, X, y):
        # TODO: Build the tree and store it in self.root
         self.root = self.build_tree(X,y)

        
    def predict_one(self, row):
        # TODO: Start from the root
        # TODO: Move left or right until reaching a leaf
        # TODO: Return the leaf value

        node = self.root

        while not node.is_leaf():
            if row[node.feature] <= node.threshold:
                node = node.left
            else:   
                node = node.right

        return node.value 

    def predict(self, X):
        # TODO: Predict each row in X

        predictions = np.zeros(len(X))
        i = 0
        for row in X:
            temp = self.predict_one(row)

            predictions[i] = temp
            i += 1
        return predictions
```

### Create an evaluation function

```python
def evaluate_model(y_true, y_pred, model_name):
    # TODO: Calculate accuracy
    # TODO: Calculate precision
    # TODO: Calculate recall
    # TODO: Calculate F1-score
    # TODO: Print the results in a nice format
    
    tp = np.sum((y_true == 1) & (y_pred == 1))
    tn = np.sum((y_true == 0) & (y_pred == 0))
    fp = np.sum((y_true == 0) & (y_pred == 1))
    fn = np.sum((y_true == 1) & (y_pred == 0))


    accuracy = (tp + tn) / (tp + tn + fp + fn)

    precision = tp / (tp + fp) 

    recall = tp / (fn + tp)

    F1_score = 2 * (precision * recall) / (precision + recall)

    print(f" Performance Metrics: {model_name}")
    print(f"------------------------------------------")
    print(f"Accuracy:  {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall:    {recall:.4f}")
    print(f"F1-Score:  {F1_score:.4f}")
    print(f"------------------------------------------")
    print(f"Confusion Breakdown ====> TP: {tp} | TN: {tn} | FP: {fp} | FN: {fn}")


```

### Logistic regression from scratch

```python
class SoftmaxRegressionScratch:
        """
        Multiclass logistic regression using softmax and gradient descent.
        """
        def __init__(self, learning_rate=0.01, n_iterations=1000):
            """
            Initialize the model.
            Args:
                learning_rate (float): step size for gradient descent.
                n_iterations (int): number of training epochs.
            """
            ## TODO: self.lr = ...
            ## TODO: self.n_iter = ...
            self.weights = None   
            self.bias = None     
            self.lr = learning_rate
            self.n_iter = n_iterations
            self.loss_history = []

        def _softmax(self, z):
            """
            Compute softmax probabilities for each sample.
            Args:
                z (np.ndarray): linear output, shape (n_samples, n_classes)
            Returns:
                np.ndarray: probabilities, shape (n_samples, n_classes)
            """
            # TODO: implement softmax with numerical stability
            exp_z = np.exp(z - np.max(z, axis = 1,keepdims=True))
            probabilites = exp_z / np.sum(exp_z,axis=1,keepdims=True)

            return probabilites

        def _one_hot(self, y, n_classes):
            """
            Convert integer labels to one-hot encoding.
            Args:
                y (np.ndarray): labels, shape (n_samples,)
                n_classes (int): number of unique classes
            Returns:
                np.ndarray: one-hot matrix, shape (n_samples, n_classes)
            """
            # TODO: implement one-hot encoding
            return np.eye(n_classes)[y]

        def fit(self, X, y):
            """
            Train the model using gradient descent.
            Args:
                X (np.ndarray): training features, shape (n_samples, n_features)
                y (np.ndarray): training labels, shape (n_samples,)
            """
            n_samples, n_features = X.shape
            n_classes = len(np.unique(y))
            
            # Initialize parameters
            # TODO: initialize self.weights (zeros) and self.bias (zeros)
            self.weights = np.zeros((n_features,n_classes))
            self.bias = np.zeros(n_classes)
            # One-hot encode labels
            # TODO: call self._one_hot(y, n_classes)
            one_hot = self._one_hot(y,n_classes)
            # Gradient descent loop
            for i in range(self.n_iter):
                # Forward pass: compute linear output z, then probabilities
                # TODO
                z = np.dot(X,self.weights) + self.bias
                probablities = self._softmax(z)
                # Compute cross-entropy loss (store in self.loss_history if desired)
                loss = -np.mean(np.sum(one_hot * np.log(probablities),axis=1))
                self.loss_history.append(loss)
                # Compute gradients
                # TODO: dw = 
                # TODO: db = 
                dz = probablities - one_hot
                dw = (1/n_samples) * np.dot(X.T,dz)
                db = (1/n_samples) * np.sum(dz,axis=0)
                # Update parameters
                # TODO: 
                self.weights -= self.lr*dw
                self.bias -= self.lr*db

                # print loss every 200 iterations
                if i % 200 == 0:
                    print(f"Iter {i}: loss = {loss:.4f}")

        def predict_proba(self, X):
            """
            Predict class probabilities for new data.
            Args:
                X (np.ndarray): features, shape (n_samples, n_features)
            Returns:
                np.ndarray: probabilities, shape (n_samples, n_classes)
            """
            # TODO: compute linear output and apply softmax
            z = np.dot(X, self.weights) + self.bias
            
            probabilities = self._softmax(z)
            
            return probabilities

        def predict(self, X):
            """
            Predict class labels for new data.
            Args:
                X (np.ndarray): features, shape (n_samples, n_features)
            Returns:
                np.ndarray: predicted labels, shape (n_samples,)
            """
            # TODO: get probabilities and return argmax
            probabilities  = self.predict_proba(X)

            predictions = np.argmax(probabilities,axis=1)

            return predictions
```

```python
k_values = [1, 3, 5, 7, 11, 21, 51]
train_accuracies = []
val_accuracies = []

for k in k_values:
    knn = KNeighborsClassifier(k=k)
    knn.fit(X_train_scaled, y_train)
    
    # Train accuracy (predict on training set)
    y_train_pred = knn.predict(X_train_scaled)
    train_acc = accuracy_score(y_train, y_train_pred)
    train_accuracies.append(train_acc)
    
    # Validation accuracy
    y_val_pred = knn.predict(X_val_scaled)
    val_acc = accuracy_score(y_val, y_val_pred)
    val_accuracies.append(val_acc)
    
    print(f"k={k:2d} -> Train Acc: {train_acc:.4f}, Val Acc: {val_acc:.4f}")
```

```python
plt.figure(figsize=(8, 5))
plt.plot(k_values, train_accuracies, marker='o', label='Train Accuracy')
plt.plot(k_values, val_accuracies, marker='s', label='Validation Accuracy')
plt.xlabel('k (number of neighbors)')
plt.ylabel('Accuracy')
plt.title('Effect of k on Model Performance (KNN)')
plt.legend()
plt.grid(True)
plt.show()
```
